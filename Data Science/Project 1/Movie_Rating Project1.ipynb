```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
df = pd.read_csv('Movie-Ratings.csv')
```

## Given Dataset

```python
df
```

## List of Top  10 movie (descending order) in terms of Budget

```python
df_des = df.sort_values('Budget (million $)',ascending = False).head(10)
df_des
```

```python
sns.set(style="whitegrid")
Descending_list = sns.barplot(x='Film',y='Budget (million $)',data = df_des)
plt.xticks(rotation=-90)
```

## List of Top 10 movies (ascending order) in terms of Budget

```python
df_asc = df.sort_values('Budget (million $)',ascending = True).head(10)
df_asc
```

```python
sns.set(style="whitegrid")
Ascending_list = sns.barplot(x='Film',y='Budget (million $)',data = df_asc)
plt.xticks(rotation=-45)
```

## Which Genre is the Best in terms of Price Budget

```python
df.groupby(['Genre'])['Budget (million $)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Genre',y = 'Budget (million $)',data=df)
```

## Genre with Highest Budget

```python
df.groupby(['Genre'])['Budget (million $)'].mean().sort_values(ascending=False).head(1)
```

## Genre with Lowest Budget

```python
df.groupby(['Genre'])['Budget (million $)'].mean().sort_values(ascending=True).head(1)
```
